```python
#5 Generate and check hypothesis around Amount Spent on different categories and response rate in different marketing campaigns
```

```python
from scipy.stats import chi2_contingency



# List of spending categories
spending_categories = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 
                       'MntSweetProducts', 'MntGoldProds']

# List of marketing campaigns
marketing_campaigns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 
                       'AcceptedCmp4', 'AcceptedCmp5']

# Iterate through each spending category and each marketing campaign
for category in spending_categories:
    for campaign in marketing_campaigns:
        # Create a contingency table
        contingency_table = pd.crosstab(merged_df[category], merged_df[campaign])
        
        # Perform chi-square test
        chi2, p, _, _ = chi2_contingency(contingency_table)
        
        # Print results
        print(f"\nChi-square test for {category} and {campaign}:")
        print(f"Chi-square value: {chi2}")
        print(f"P-value: {p}")

```

```python
#8. Can you test the hypothesis that recent customers complain less in general compared to older customers?
```

```python

from scipy.stats import chi2_contingency

# Assuming df is your dataframe
# Convert 'Dt_Customer' to datetime format
merged_df['Dt_Customer'] = pd.to_datetime(merged_df['Dt_Customer'])

# Define recent and older customers
current_date = pd.to_datetime('today')  # Assuming today's date as the current date
six_months_ago = current_date - pd.DateOffset(months=6)

recent_customers = merged_df[merged_df['Dt_Customer'] >= six_months_ago]
older_customers = merged_df[merged_df['Dt_Customer'] < six_months_ago]

# Calculate proportions of customers who complained in each group
recent_complain_proportion = recent_customers['Complain'].mean()
older_complain_proportion = older_customers['Complain'].mean()

print(f"Proportion of recent customers who complained: {recent_complain_proportion}")
print(f"Proportion of older customers who complained: {older_complain_proportion}")

# Perform chi-square test
contingency_table = pd.crosstab(merged_df['Dt_Customer'] >= six_months_ago, merged_df['Complain'])
chi2, p_value, _, _ = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in complaint proportions between recent and older customers.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in complaint proportions between recent and older customers.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Calculate the mean income and mean age for both responders and non-responders
mean_income = merged_df.groupby('Response')['Income'].mean()
mean_age = merged_df.groupby('Response')['Age'].mean()

# Calculate the demographic distribution, marital status distribution, and country distributio
education_distribution = merged_df.groupby(['Response', 'Education']).size().unstack()
marital_status_distribution = merged_df.groupby(['Response', 'Marital_Status']).size().unstack()
country_distribution = merged_df.groupby(['Response', 'Country']).size().unstack()

# Plotting the charts
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
# Mean Income
mean_income.plot(kind='bar', color=['pink', 'skyblue'], ax=axes[0, 0])
axes[0, 0].set_title('Mean Income for Responders vs. Non-Responders')
axes[0, 0].set_xlabel('Response')
axes[0, 0].set_ylabel('Mean Income')

# Mean Age
mean_age.plot(kind='bar', color=['purple', 'green'], ax=axes[0, 1])
axes[0, 1].set_title('Mean Age for Responders vs. Non-Responders')
axes[0, 1].set_xlabel('Response')
axes[0, 1].set_ylabel('Mean Age')

# Education Distribution
education_distribution.plot(kind='bar', stacked=True, ax=axes[1, 0])
axes[1, 0].set_title('Education Distribution for Responders vs. Non-Responders')
axes[1, 0].set_xlabel('Education')
axes[1, 0].set_ylabel('Count')

# Marital Status Distribution
marital_status_distribution.plot(kind='bar', stacked=True, ax=axes[1, 1])
axes[1, 1].set_title('Marital Status Distribution for Responders vs. Non-Responders')
axes[1, 1].set_xlabel('Marital Status')
axes[1, 1].set_ylabel('Count')

# Country Distribution
country_distribution.plot(kind='bar', stacked=True, ax=axes[2, 0])
axes[2, 0].set_title('Country Distribution for Responders vs. Non-Responders')
axes[2, 0].set_xlabel('Country')
axes[2, 0].set_ylabel('Count')

# Remove the empty subplot
fig.delaxes(axes[2, 1])

plt.tight_layout()
plt.show()
```
